```python
# there is problem in society column that
# 1) some names are in capital letter and some name in small letter
# 2) some name has rating also in thier name
import re
df['society'] = df['society'].apply(lambda name : re.sub(r'\d+(\.\d+)?\s?★' , '' , str(name)).strip()).str.lower()
```

```python
df['society'] = df['society'].str.replace('nan','independent')
```

```python
df[df['price']=='Price on Request']
```

```python
#1) some column has Price on Request insteed of price means not specified price
# so we will remove these rows beacuse price is our target column in our model
df = df[df['price']!='Price on Request']
```

```python
# price
# 1) some prices in column and some price in lakh
def treet_price(x):
  if type(x) == float:
    return x
  else:
    if x[1] == 'Lac':
      return round(float(x[0])/100,2)
    else:
      return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treet_price)
```

```python
df.head()
```

```python
df['price_per_sqft'].value_counts()
```

```python
# price_per_sqft has ₹ , ',' , / , sq.ft we have to clean it

df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.split(' ').str.get(1).str.replace(',','').str.strip().astype(float)
```
